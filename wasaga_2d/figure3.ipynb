```python
import matplotlib.pyplot as plt

# Define distances and observed SRP masses
distances = [50, 70, 90]
observed_srp = [12730, 12500, 16210]
observed_residual_srp = [6700, 3200, 1000]

# Simulated SRP mass calculation, adjusted by a factor
simulated_srp = [8013, 9601, 4711]

# Plotting observed SRP
plt.plot(distances, observed_srp, 'r-', label='Observed SRP')
plt.scatter(distances, observed_srp, c='r')



# Plotting observed residual SRP
plt.plot(distances, observed_residual_srp, 'r--', label='Residual SRP')
plt.scatter(distances, observed_residual_srp, c='r')

# Shaded area between observed SRP and residual SRP
plt.fill_between(distances, observed_srp, observed_residual_srp, color='lightgrey', label='Target SRP Range',zorder=0,alpha=0.5)

# Plotting simulated SRP
plt.plot(distances, simulated_srp, 'r:', label='Simulated SRP')
plt.scatter(distances, simulated_srp, c='r')



# Labeling axes
plt.ylabel('SRP Mass (mg)')
plt.xlabel('Distance (m)')
plt.rcParams.update({'font.size': 10})

# Adding vertical lines
for x in distances:
    plt.axvline(x=x, alpha=0.5, c='grey', linestyle='--')

# Adding a legend
plt.legend(bbox_to_anchor=(0.27, 0.45), fontsize=8)

# Setting the axis limits
plt.ylim(-1000, 19000)
plt.xlim(30, 100)

# Setting the figure size
plt.gcf().set_size_inches(7, 2.5)

# Show plot
plt.show()
```
