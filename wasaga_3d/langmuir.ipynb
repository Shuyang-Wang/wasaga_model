```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_q(K_L, Q, C):
    """
    Calculate the solid-liquid partitioning using the Langmuir isotherm.

    Parameters:
    - K_L: Langmuir adsorption constant (L/mg)
    - Q: Maximum adsorption capacity (mg/g)
    - C: Equilibrium concentration of solute in liquid phase (mg/L)

    Returns:
    - q: Amount of solute adsorbed per unit mass of solid (mg/g)
    """
    q = (Q * K_L * C) / (1 + K_L * C)
    return q

def plot_q_vs_C(C_values, q_values):
    """
    Plot q (amount adsorbed) vs. C (equilibrium concentration).

    Parameters:
    - C_values: Array of equilibrium concentrations (mg/L)
    - q_values: Array of corresponding q values (mg/g)
    """
    plt.figure(figsize=(8, 6))
    plt.plot(C_values, q_values, 'b-', label='Langmuir Isotherm')
    plt.title('Langmuir Isotherm: q vs. C')
    plt.xlabel('Equilibrium Concentration, C (mg/L)')
    plt.ylabel('Amount Adsorbed, q (mg/g)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
K_L = 0.2  # Example Langmuir constant, L/mg
Q = 100*1000    # Example maximum adsorption capacity, mg/g

# Equilibrium concentrations (mg/L)
C_values = np.linspace(0, 5000, 200)  # Generates 100 values from 0 to 10

# Calculate q for each C
q_values = [calculate_q(K_L, Q, C) for C in C_values]

# Plot q vs. C
plot_q_vs_C(C_values, q_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt



def calculate_percentages(K_L, Q, C):
    """
    Calculate the percentage of phosphorus in aqueous and solid phases.

    Parameters:
    - K_L: Langmuir adsorption constant (L/mg)
    - Q: Maximum adsorption capacity (mg/g)
    - C: Equilibrium concentration of solute in liquid phase (mg/L)

    Returns:
    - P_aqueous: Percentage of phosphorus in the aqueous phase
    - P_solid: Percentage of phosphorus in the solid phase
    """
    q = calculate_q(K_L, Q, C)
    total_P = q + C
    P_aqueous = (C / total_P) * 100
    P_solid = (q / total_P) * 100
    return P_aqueous, P_solid

def plot_percentages(C_values, P_aqueous_values, P_solid_values):
    """
    Plot the percentage of phosphorus in aqueous and solid phases vs. C.

    Parameters:
    - C_values: Array of equilibrium concentrations (mg/L)
    - P_aqueous_values: Array of corresponding aqueous phase percentages
    - P_solid_values: Array of corresponding solid phase percentages
    """
    plt.figure(figsize=(8, 6))
    plt.plot(C_values, P_aqueous_values, 'r-', label='Aqueous Phase')
    plt.plot(C_values, P_solid_values, 'g-', label='Solid Phase')
    plt.title('Phosphorus Distribution: Aqueous vs. Solid Phases')
    plt.xlabel('Equilibrium Concentration, C (mg/L)')
    plt.ylabel('Percentage (%)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
K_L = 0.2  # Example Langmuir constant, L/mg
Q = 0.1   # Example maximum adsorption capacity, mg/g

# Equilibrium concentrations (mg/L)
C_values = np.linspace(0, 10000, 1000)  # Generates 1000 values from 0 to 100

# Calculate percentages for each C
P_aqueous_values = []
P_solid_values = []

for C in C_values:
    P_aqueous, P_solid = calculate_percentages(K_L, Q, C)
    P_aqueous_values.append(P_aqueous)
    P_solid_values.append(P_solid)

# Plot percentages vs. C
plot_percentages(C_values, P_aqueous_values, P_solid_values)
```

```python
def show_percentages(K_L, Q, C):
    """
    Show the percentage of phosphorus in aqueous and solid phases at a single concentration.

    Parameters:
    - K_L: Langmuir adsorption constant (L/mg)
    - Q: Maximum adsorption capacity (mg/g)
    - C: Equilibrium concentration of solute in liquid phase (mg/L)

    Prints:
    - Percentage of phosphorus in the aqueous phase
    - Percentage of phosphorus in the solid phase
    """
    P_aqueous, P_solid = calculate_percentages(K_L, Q, C)
    print(f"At equilibrium concentration {C} mg/L:")
    print(f"Percentage of phosphorus in the aqueous phase: {P_aqueous:.2f}%")
    print(f"Percentage of phosphorus in the solid phase: {P_solid:.2f}%")

# Example usage
K_L = 0.2  # Example Langmuir constant, L/mg
Q = 0.1   # Example maximum adsorption capacity, mg/g
C = 0.1    # Example equilibrium concentration, mg/L

# Show percentages at the given concentration
show_percentages(K_L, Q, C)
```
